```python
from os import urandom
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_port():
    return urandom(1)[0] % 3 + 1

GAMES = 1000

game = 0
win = 0

# 2xGAMES matrix
history = np.array([0.0] * GAMES)

for game in range(GAMES):
    car_port = random_port()
    chosen_port = random_port()
    # if car_port == chosen_port: now AI open one of the 2 other ports and you change -> you lose
    # if car_port != chosen_port: now AI open the wrong port and the other port not chosen and not open is the right port -> you win
    win += car_port != chosen_port
    history[game] = win / (game+1)
```

```python
# calc regression line of history
slope, intercept = np.polyfit(range(GAMES), history, 1)

# show temporal evolution
plt.plot(history)
plt.plot(range(GAMES), slope * range(GAMES) + intercept, color="green")
# show in red asyntotic probability = 66%
plt.axhline(2/3, color='red')
```
