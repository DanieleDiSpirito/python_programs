```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.array(np.sin(np.linspace(0, 10 * np.pi, 1000)))

plt.rcParams['figure.figsize'] = [10, 7]

ITER = 100_000
j = 0

for i in range(2, ITER+1):
    A = np.random.rand()
    W = (np.random.rand() - 0.5) * 5
    C = (np.random.rand() - 0.5) * 2
    g = np.array(A * np.sin(W * np.linspace(0, 10 * np.pi, 1000) + C * np.pi))
    f += g
    if i % 10**j == 0:
        print(f"Iteration {i}")
        plt.plot(f, color='blue')
        j += 1
    
    plt.show()
```

```python
LEN = 100
ITER = 30
matrix = np.matrix([[0] * LEN for _ in range(LEN)] , dtype=np.float64)

plt.rcParams['figure.figsize'] = [10, 7]

for x in range(LEN):
    for y in range(LEN):
        deltaX = np.random.randint(-5, 5+1)
        deltaY = np.random.randint(-5, 5+1)
        if x + deltaX >= 0 and x + deltaX < LEN and y + deltaY >= 0 and y + deltaY < LEN:
            matrix[x + deltaX, y + deltaY] += 1

matrix0 = matrix.copy()

for i in range(ITER):
    new_matrix = np.matrix([[0] * LEN for _ in range(LEN)] , dtype=np.float64)
    if i % 5 == 0:
        print(f"Iteration #{i}")
        plt.imshow(matrix, cmap='hot', interpolation='nearest')
        plt.colorbar()
        plt.show()
    for x in range(LEN):
        for y in range(LEN):
            n_neighbors = 0
            for deltaX in range(-1, 2):
                for deltaY in range(-1, 2):
                    if x + deltaX >= 0 and x + deltaX < LEN and y + deltaY >= 0 and y + deltaY < LEN and (deltaX != 0 or deltaY != 0):
                        new_matrix[x, y] += matrix[x + deltaX, y + deltaY]
                        n_neighbors += 1
            new_matrix[x, y] /= n_neighbors
    matrix = new_matrix

matrix1 = matrix.copy()

print(f"Iteration #{ITER}")
plt.imshow(matrix, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
# show 3d plot of matrix0 and matrix1
x = np.arange(matrix0.shape[0])
y = np.arange(matrix0.shape[1])
X, Y = np.meshgrid(x, y)

plt.rcParams['figure.figsize'] = [25, 20]
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, matrix0, cmap='hot')
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X, Y, matrix1, cmap='hot')
plt.show()
```
