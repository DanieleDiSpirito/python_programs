```python
from random import random
from math import pi as PI

def distance(P, Q):
    return ((P[0] - Q[0]) ** 2 + (P[1] - Q[1]) ** 2) ** 0.5

in_circle = 0
in_square = 0

for _ in range(10_000_000):
    P = (random(), random())
    if distance(P, (1, 1)) < 1:
        in_circle += 1
    in_square += 1

pi: float = 4 * in_circle / in_square
print(f"{pi = }")
print(f"{PI = }")
print(f"Error: {100*abs(pi - PI):.5}%")
```

```python
from random import random
import matplotlib.pyplot as plt

def distance(P, Q):
    return ((P[0] - Q[0]) ** 2 + (P[1] - Q[1]) ** 2) ** 0.5

in_circle = 0
in_square = 0

points = []

for _ in range(10_000):
    P = (random()*2, random()*2)
    if distance(P, (1, 1)) < 1:
        in_circle += 1
    in_square += 1
    points.append(P)

plt.figure(figsize=(8,8))
plt.scatter([point[0] for point in points], [point[1] for point in points], s=0.5, color='blue')

circle = plt.Circle((1, 1), 1, fill=False, color='red')
plt.gca().add_patch(circle)
plt.axis('equal')
plt.title('Monte Carlo Simulation for Pi Estimation')
plt.show()
```

```python
from random import gauss
import matplotlib.pyplot as plt

def distance(P, Q):
    return ((P[0] - Q[0]) ** 2 + (P[1] - Q[1]) ** 2) ** 0.5

in_circle = 0
in_square = 0

points = []

for _ in range(10_000):
    P = (gauss(1, 0.3), gauss(1, 0.3))
    if distance(P, (1, 1)) < 1:
        in_circle += 1
    in_square += 1
    points.append(P)

plt.figure(figsize=(8,8))
plt.scatter([point[0] for point in points], [point[1] for point in points], s=0.5, color='blue')

circle = plt.Circle((1, 1), 1, fill=False, color='red')
plt.gca().add_patch(circle)
plt.axis('equal')
plt.title('Gaussian distribution')
plt.show()
```
