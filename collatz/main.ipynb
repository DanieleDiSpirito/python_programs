```python
import matplotlib.pyplot as plt

def collatz(number):
    sequence = []
    while number != 1:
        sequence.append(number)
        if number % 2 == 0:
            number = number // 2
        else:
            number = 3 * number + 1
    sequence.append(1)  # Aggiungi l'ultimo termine della sequenza, che Ã¨ 1
    return sequence

# Esempio di utilizzo della funzione
for i in range(1, 10_000):
    plt.plot(collatz(i))

plt.show()
```
