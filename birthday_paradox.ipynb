```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PEOPLE = 100
DAY = 365
```

```python
prob = np.array([1] * PEOPLE, float)

for i in range(1, PEOPLE):
    prob[i] = prob[i-1] * (365-i)/365

plt.plot(100 - 100*prob, 'k-')
plt.axhline(y = 50, color='r', linestyle='-')
plt.axvline(x = 22, color='b', linestyle='-')
plt.show()
```
